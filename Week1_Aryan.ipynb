````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

for i in range(len(eigenvalues)):
    λ = eigenvalues[i]
    v = eigenvectors[:, i]
    assert np.allclose(A @ v, λ * v), "Eigenvector equation does not hold!"


# Plot
is_valid = np.allclose(A @ eigenvectors, eigenvectors @ np.diag(eigenvalues))
if is_valid:
  print("valid eigenvectors")
else:
  print("Invalid eigenvectors")

print("4.3 passed ✓")
````

**Geometric Explanation (4.3):**

An eigenvector is a unique vector whose direction remains unchanged
, when transformation is applied. and eigenvalues are scalar factor that causes a certain vector to be stretched, compressed, or flipped.



#### **Formula:**
 $A\vec{v} = λ\vec{v}$
#### Where :


*   "A" is transformation Matrix
* λ is eiganvalue
* ⃗v is eiganvector

---

The eigenvalue λ functions as a scaling factor. It indicates precisely what the
matrix is doing to that particular axis of space, depending on its value:


1.  Stretch (λ > 1): Any vector along the line of that eigenvector doubles in length if λ = 2. Space is being pulled outward in this direction by the matrix.
2.  Squish (0 < λ < 1): Vectors along this line are chopped in half if λ = 0.5. Along this axis, the matrix is squeezing space inward.


### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X,full_matrices = False)

X_reconstructed = U @ np.diag(S) @ Vt
X_approx        = U[:, :1] @ np.diag(S[:1]) @ Vt[:1, :]

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

SVD decomposes a matrix X into X = U · S · Vᵀ

* Vᵀ (V-transpose): Rows are the right singular vectors - these are the principal component directions (eigenvectors of XᵀX). Each row points in a direction of variance in the original feature space.
* S (Diagonal matrix): Contains singular values (σ₁ ≥ σ₂ ≥ ...) - their squares divided by (n−1) give the variance explained by each principal component.
* U: Contains left singular vectors — the coordinates of data points in the new PC space (scores).
---
Relation to PCA:
* When you run PCA via SVD on the centered matrix X, the columns of V (rows of Vᵀ) are exactly the principal component axes (same as eigenvectors of the covariance matrix).
* V (or Vᵀ) contains the directions of maximum variance.
* U gives where each sample lands; S tells how much variance each direction captures; V tells which direction in feature space.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff =  adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    epsilon = 1e-10
    observed_hist, _ = np.histogram(actual, bins=bins)
    expected_hist, _ = np.histogram(expected, bins=bins)
    observed_hist = observed_hist / observed_hist.sum()
    expected_hist = expected_hist / expected_hist.sum()
    psi_scores = (observed_hist - expected_hist) * np.log(
        (observed_hist + epsilon) / (expected_hist + epsilon)
    )
    return psi_scores.sum()

    pass

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=20, alpha=0.5, label='Training')
ax.hist(drift_dist, bins=20, alpha=0.5, label='Production')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** The statistical properties of Target variable, which the model is trying to predict, changes over time.
- **Covariate drift:** The change in the distribution of one or more independent variables or input variables of a dataset.
- **PSI < 0.1:** It means stable shift, the distribution has not changed , so no action is required.
- **PSI 0.1–0.2:** It means modrate shift, the distribution is shifting. Monitor performance closely and prepare for an upcoming model update.
- **PSI > 0.2:** It meand a major shift, significant data drift detected. Requires immediate intervention or automated retraining.
- **Retraining trigger example:** Every 24 hours, an automatic monitoring script uses the Population Stability Index (PSI) to assess incoming production data. The PSI value exceeds the > 0.2 criterion if a crucial input feature drifts significantly, such as an income distribution moving from a mean of 50 to 65. In order to maintain the accuracy and objectivity of the model's predictions, this automatically initiates a process to retrain the model using the most recent data.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = pd.Series(red/total).mean()
p_blue  = pd.Series(blue/total).mean()
p_green = pd.Series(green/total).mean()

# Joint: P(red first AND blue second)
p_red_then_blue = p_red * p_blue

# Conditional: P(blue second | red first)
p_blue_given_red = p_red_then_blue / p_red

# Independence check: are P(B|A) == P(B) ?
independent = p_blue_given_red == p_blue

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.zeros(n_samples)

for i in range(n_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means[i] = sample.mean()

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
ax.hist(sample_means, bins=30, density=True, alpha=0.5, label='Sample Means')
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> The Central Limit Theorem (CLT) is fundamental because it ensures that, regardless of how chaotic or skewed the original data's shape appears, the distribution of its sample mean will naturally form a bell curve as a sample size increases. Data scientists can confidently compute margins of error, create confidence intervals, and forecast the behavior of averages across large datasets thanks to its predictable behavior.

> The CLT directly supports the assumption of normalcy in traditional statistical testing. By confirming that the test statistics will behave reliably even if the underlying population data itself is non-normal, this fundamental premise supports numerous common inferences, including the one-sample t-test you performed earlier.
